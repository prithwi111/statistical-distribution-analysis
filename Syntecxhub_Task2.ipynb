```python
#Import Libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#It changes the background style of your plots.
#Make the plot look cleaner and easier to reads
sns.set_style("whitegrid")
```

```python
#Load Dataset by using the Seaborn
data=sns.load_dataset("tips")
```

```python
#fetch some initial Records of data
data.head()
```

```python
#It gives summary statistics of numerical columns in a dataset.
data.describe()
```

```python
#It gives a structural summary of the dataset.
data.info()
```

```python
#Histogram (Distribution Overview)
plt.figure(figsize=(8,5))
sns.histplot(data['total_bill'],bins=20,kde=False)
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.savefig("histogram.png")
plt.show()
plt.close()
```

```python
#KDE Plot (Density Curve)
plt.figure(figsize=(8,5))
sns.kdeplot(data['total_bill'],fill=True)
plt.title("KDE Plot of Total Bill")
plt.savefig("Kdeplot.png")
plt.show()
```

```python
#Boxplot (Outlier Detection)
plt.figure(figsize=(8,5))
sns.boxplot(x=data['total_bill'])
plt.title("Boxplot of Total Bill")
plt.savefig("Boxplot.png")
plt.show()
```

```python
#Compare Groups (Smoker vs Non-Smoker)
plt.figure(figsize=(8,5))
sns.boxplot(x='smoker', y='total_bill', data=data)
plt.title("Total Bill Distribution: Smoker vs Non-Smoker")
plt.savefig("Smoker vs Non-Smoker.png")
plt.show()
```

```python
#Histogram Comparison by Group
plt.figure(figsize=(8,5))
sns.histplot(data=data, x='total_bill', hue='smoker', kde=True)
plt.title("Distribution Comparison by Smoker Status")
plt.savefig("Comparison_bySmoker_Status.png")
plt.show()
```
